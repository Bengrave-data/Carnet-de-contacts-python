## 1. Initialisation du carnet

```python
# Mini projet : Carnet de contacts

carnet = [
    {"nom": "Cartman", "prenom": "Eric", "telephone": "0456-96-17-45", "ville": "South Park - Colorado"},
    {"nom": "McCormick", "prenom": "Kenny", "telephone": "0491-22-86-07", "ville": "South Park - Colorado"},
    {"nom": "Marsh", "prenom": "Stan", "telephone": "0458-86-87-85", "ville": "South Park - Colorado"},
    {"nom": "Broflovski", "prenom": "Kyle", "telephone": "0486-98-18-48", "ville": "South Park - Colorado"},
]

print("Carnet initialisé avec", len(carnet), "contacts")

```

## 2. Fonction d'affichage

```python
# Fonction pour afficher tout le carnet

def afficher_carnet(carnet):
  print("📇 +++ Mon carnet de contacts +++ 📇")
  print("Total :", len(carnet), "contacts")
  print()

for contact in carnet:
  print("🇺🇸", contact["nom"], contact["prenom"])
  print("   📞", contact["telephone"])
  print("   🏙️", contact["ville"])
  print("+++")

# Test
afficher_carnet(carnet)
```

## 3. Fonction d'ajout

```python
def ajouter_contact(carnet, nom, prenom, telephone, ville):
    nouveau_contact = {
        "nom": nom,
        "prenom": prenom,
        "telephone": telephone,
        "ville": ville
    }
    carnet.append(nouveau_contact)
    print("✅ Contact ajouté :", prenom, nom)

# Test ajouttons Chef
ajouter_contact(carnet, "Hayes", "Chef", "0488-98-08-68", "South Park - Colorado")
afficher_carnet(carnet)
```

```python
# Test rapide

ajouter_contact(carnet, "Garrison", "Mrs", "0454-99-06-69", "South Park - Colorado")
ajouter_contact(carnet, "Black", "Token", "0431-32-36-37", "South Park - Colorado")

afficher_carnet(carnet)
```

## 4. Fonction de recherche

```python
# Fonction pour rechercher un contact par son nom
def rechercher_contact(carnet, nom_recherche):
    trouve = False

    for contact in carnet:
        if contact["nom"] == nom_recherche:
            print("✅ Contact trouvé :")
            print("  ", contact["prenom"], contact["nom"])
            print("  📞", contact["telephone"])
            print("  📊", contact["ville"])
            trouve = True

    if not trouve:
        print("❌ Aucun contact trouvé avec le nom :", nom_recherche)

# Test 1 : un contact qui existe
rechercher_contact(carnet, "McCormick")

print()

# Test 2 : un contact qui n'existe pas
rechercher_contact(carnet, "Trump")
```
